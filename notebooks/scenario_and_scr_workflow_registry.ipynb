```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from model.config import (
    AssumptionConfig,
    ScenarioConfig,
    CorrelationConfig,
    CapitalWorkflowConfig
)

from model.capital.capital_workflow import (
    run_capital_workflow
)
```

```python
assumption_config = AssumptionConfig(
    values={
        "mortality_table_path":
            "../data/mortality_tables/ons_mortality.csv",

        "mortality_parameter_path":
            "../data/mortality_parameters/smoker_multipliers.csv",

        "yield_curve_path":
            "../data/yield_curves/sonia_spot_rates.csv",

        "lapse_table_path":
            "../data/lapse_tables/lapse_rates.csv",

        "expense_table_path":
            "../data/expense_tables/expense_rates.csv"
    }
)

scenario_config = ScenarioConfig(
    scenario_path=
        "../data/scenarios/standard_formula.csv"
)

correlation_config = CorrelationConfig(
    life_correlation_path=
        "../data/correlations/life/life_correlation_matrix.csv",

    market_correlation_path=
        "../data/correlations/market/market_correlation_matrix.csv",

    bscr_correlation_path=
        "../data/correlations/bscr/bscr_correlation_matrix.csv"
)

workflow_config = CapitalWorkflowConfig(
    portfolio_path=
        "../data/portfolios/sample_portfolio.csv",

    assumption_config=
        assumption_config,

    scenario_config=
        scenario_config,

    correlation_config=
        correlation_config,

    return_breakdown=True
)
```

```python
result = run_capital_workflow(
    workflow_config
)

print("=" * 60)
print("BASE VALUATION")
print("=" * 60)

print(
    f"BEL: "
    f"{result.base_result.best_estimate_liability:,.2f}"
)

print()

print("=" * 60)
print("UNIVARIATE SCRS")
print("=" * 60)

for scr in sorted(
    result.scr_results,
    key=lambda x: (
        x.aggregation_category,
        x.risk_type
    )
):

    print(
        f"{scr.aggregation_category:<10}"
        f"{scr.risk_type:<15}"
        f"{scr.scr:>15,.2f}"
    )

print()

print("=" * 60)
print("LIFE SCR")
print("=" * 60)

print(
    f"Gross: "
    f"{result.life_scr.gross_scr:,.2f}"
)

print(
    f"Diversified: "
    f"{result.life_scr.diversified_scr:,.2f}"
)

print()

print("=" * 60)
print("MARKET SCR")
print("=" * 60)

print(
    f"Gross: "
    f"{result.market_scr.gross_scr:,.2f}"
)

print(
    f"Diversified: "
    f"{result.market_scr.diversified_scr:,.2f}"
)

print()

print("=" * 60)
print("BSCR")
print("=" * 60)

print(
    f"Gross: "
    f"{result.bscr.gross_scr:,.2f}"
)

print(
    f"Diversified: "
    f"{result.bscr.diversified_scr:,.2f}"
)
```
